```python
#to check which one is method and which one is attribute we can use the code bellow:
for item in dir(b):
    if not item.startswith("__"):
        val = getattr(b, item)
        if callable(val):
            print(f"{item} → Method")
        else:
            print(f"{item} → Attribute")



    
```
