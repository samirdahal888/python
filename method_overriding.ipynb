## method overriding


```python
class Employee:
    def __init__(self,department,salary):
        
        self.department = department
        self.salary = salary
    
    def Get_salary(self):
        return self.salary
    
class Salesofficer(Employee):
    def __init__(self, department, salary,name): #use old code  and add some new to each class specific 
        super().__init__(department, salary)
        self.name = name

    def Get_salary(self): # gets the old method as well and add new functionality as well , code reuseability and can extend
        base_salary =  super().Get_salary()
        total_salary  =  base_salary + 1000
        return total_salary
```

```python
s= Salesofficer('sales',200,'samir')
```
